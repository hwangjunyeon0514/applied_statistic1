```python
import cvxpy as cp
```

```python
x1 = cp.Variable(nonneg=True) # the number of sheets daily produced
x2 = cp.Variable(nonneg=True) # the number of bars daily produced
```

```python
obj = cp.Maximize(40*x1 + 35*x2) # maximize the profit, 목적 함수 정의
```

```python
# 제약식 정의
constraints = [x1 <= 550,
               x2 <= 560,
               6*x1 + 8*x2 <= 4800]
```

```python
# 최적화 모형 구성
prob = cp.Problem(obj, constraints)
```

```python
# 디폴트 솔버로 최적화 모형 계산
prob.solve()
```

```python
# 결과 확인
print('최적해 상태: ', prob.status)
print('최적값: ', prob.value)
print('최적해: ', x1.value, x2.value)
```
